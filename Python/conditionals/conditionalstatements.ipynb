```python
## If Statement

age = 20

if age >=18 :
    # If statement excutes a block of code if the condition is true.
    print("You are allowed to vote in elections")
```

```python
## Else Statement

## The Else statement excutes a block of code if the condition in the if statement is not true.

age = 16

if age >= 18 :
    print("You are allowed to vote in elections")
else :
    print("You are not allowed to vote in elections")



```

```python
## Elif Statement

## The Elif statement excutes a block of code if the condition in the if statement is not true.
## The elif statement is used to check multiple conditions. It stands for else if.

age = 20

if age < 13 :
    print("You are a child")
elif age<18 :
    print("You are a teenager")
else :
    print("You are an adult")



```

#### Nested Conditional Statements

- You can place one or more if,elif or else statements inside another if ,elif or else statement inside another if,elif or else statement to created nested conditional statements.

```python
## Number (even,odd and negative)

num = int(input("Enter the number"))

if num>=0:
    print("The number is Positive")
    if num%2 == 0 :
        print("The number is even")
    else:
        print("The number is odd")

else:
    print("The number is zero or negative")
```

#### Pratical Example

```python
## Determine if a year is a leap year using nested conditional statement

year = int(input("Enter the year"))

if year%4 == 0 :
    if year%100 == 0:
        if year%400 == 0 :
            print(year,"is a leap year")

        else:
            print(year,"is not a leap year")
    
    else:
        print(year,"is a leap year")

else :
    print(year,"is a leap year ")
```

```python
## Simple Calculator

#Enter operation(+,-,*,/)

# Take user input
num1 = float(input("Enter first number"))
num2 = float(input("Enter second number"))

operation = input("enter operation(+,-,*,/)")

if operation == "+":
    result = num1+ num2
elif operation == "-":
    result = num1 - num2
elif operation == "*":
    result = num1 * num2
elif operation == "/":
    if num2 != 0:
        result = num1/num2
    else:
        result = "Error! Division by zero"
else:
    result = "Invalid Operation"

print("result", result)
```
