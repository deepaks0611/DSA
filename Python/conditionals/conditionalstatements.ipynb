```python
## If Statement

age = 20

if age >=18 :
    # If statement excutes a block of code if the condition is true.
    print("You are allowed to vote in elections")
```

```python
## Else Statement

## The Else statement excutes a block of code if the condition in the if statement is not true.

age = 16

if age >= 18 :
    print("You are allowed to vote in elections")
else :
    print("You are not allowed to vote in elections")



```

```python
## Elif Statement

## The Elif statement excutes a block of code if the condition in the if statement is not true.
## The elif statement is used to check multiple conditions. It stands for else if.

age = 20

if age < 13 :
    print("You are a child")
elif age<18 :
    print("You are a teenager")
else :
    print("You are an adult")



```
