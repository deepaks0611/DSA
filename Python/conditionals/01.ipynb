```python
x = int(input("Enter the number"))

if x>15 : ## false hone ke baad else run hoga
    print("Number is greater than 15")
else:
    print("Number is less than 15")
    
```

```python
x = int(input("Enter the number"))

if x>15 :
    print("Number is greater than 15")
elif x==15: ## if false hone ke baad elif run hoga when multiple condition lagne wala hota h tab elif use karte h
    print("Number is equal to 15")
else:
    print("Number is less than 15")
    
```
