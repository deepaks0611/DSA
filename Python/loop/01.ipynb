##### Loops
- Repeat multiple times

1. For Loop
2. While Loop

#### For Loop

```python
fruits = ["apple","banana","cherry","grapes"]

for fruit in fruits :
    print(fruit)
```

```python
name = "Leo Messi"

for char in name :
    print(char)    
```

```python
name = "Leo Messi"

for char in name :
    print(char, end = " ")
    
    
    
```

```python
range(10)  # range = n -> n-1 tak print hoga

for i in range(10) :
    print(i)
```

```python
for i in range (2,7):
    print(i)
```

```python
for i in range (2,10,2): # third argument (2) is step size
    print(i)
    
```

##### Range
- range(n) -> n element -> [0,1,2,3,...........,n-1]
- range(a,b) -> [a,a+1,a+2,.............,b-1]
- range(a,b,c) -> [a,a+c,a+2c,........,till(>b)]

```python
for i in range(10,0,-1) : # +ve sign means left to right [1,2,3,....] or -ve sign means right to left [10,9,8,......]
    print(i)
```

```python
for i in range(12,0,-3) :
    print(i)
```
